```python
#Importing all our libraries for this data cleaning project

import pandas as pd
from pandasql import sqldf
pysqldf = lambda q : sqldf(q, globals())
```

```python
#Import dataset

housing = pd.read_csv("Nashville Housing Data for Data Cleaning.csv")
```

```python
#Previewing your dataset

pysqldf("SELECT * FROM housing")
```

```python
housing.info()
```

```python
#Standardize our saledate format

housing['SaleDate'] = pd.to_datetime(housing['SaleDate'])
```

```python
housing.info()
```

```python
#Populating the null values within our property address

pysqldf("SELECT * FROM housing WHERE PropertyAddress IS NULL ORDER BY ParcelID")
```

```python
Addresscom = housing['PropertyAddress'].str.split(",", n = 1, expand = True)
```

```python
housing["Address"] = Addresscom[0]
housing["City"] = Addresscom[1]
```

```python
housing.drop(columns = ["PropertyAddress"], inplace = True)
```

```python
owneraddress = housing['OwnerAddress'].str.split(",", n = 1, expand = True)
```

```python
housing["OwnerAddress"] = owneraddress[0]
housing["OwnerCityState"] = owneraddress[1]
```

```python
housing.drop(columns = ["OwnerCity"], inplace = True)
housing.drop(columns = ["OwnerAddress"], inplace = True)
```

```python
owneraddress = housing['OwnerCityState'].str.split(",", n = 1, expand = True)
```

```python
housing["OwnerAdd"] = owneraddress[0]
housing["OwnerState"] = owneraddress[1]
```

```python
housing.drop(columns = ["OwnerCityState"], inplace = True)
housing.drop(columns = ["OwnerAdd"], inplace = True)
```

```python
pysqldf("SELECT * FROM housing")
```

```python
pysqldf("SELECT DISTINCT(SoldAsVacant) FROM housing")
```

```python
#Changing Y and N to Yes and No

pysqldf("SELECT SoldAsVacant, CASE WHEN SoldAsVacant = 'Y' THEN 'Yes' WHEN SoldAsVacant = 'N' THEN 'No' ELSE SoldAsVacant END FROM housing")
```

```python
housing["SoldAsVacant"] = housing["SoldAsVacant"].replace("Y", "Yes")
housing["SoldAsVacant"] = housing["SoldAsVacant"].replace("N", "No")
```

```python
pysqldf("SELECT DISTINCT(SoldAsVacant) FROM housing")
```

```python
#Removing duplicates

housing.drop_duplicates(subset = None, keep = 'first', inplace = True)
```

```python
#Preview cleaned data

pysqldf("SELECT * FROM housing")
```

```python
#Export data

housing.to_csv('Processed.csv')
```
